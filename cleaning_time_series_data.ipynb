### 1. Cleaning non-countries list

> The data files contain non countries list in Country_Regio & Province_State

```python
import pandas as pd
import numpy as np
df=pd.read_csv("data/time_series_covid19_confirmed_global.csv")
df = df[df['Country/Region'].notna()].copy()
# List of known non-country entries in the JHU data
non_countries = [
    "Diamond Princess", "Grand Princess", "MS Zaandam", "Recovered",
    "Holy See", "Summer Olympics 2020", "Winter Olympics 2022", "Kosovo", 
    "West Bank and Gaza", "Burma", "Micronesia",
    "Congo (Kinshasa)", "Congo (Brazzaville)", "Eswatini", "Korea, South",
    "Laos", "North Macedonia", "Timor-Leste", "Palestine",
    "Mainland China", "occupied Palestinian territory"
]

# Filter the dataframe to keep only valid countries
df_cleaned = df[~df['Country/Region'].isin(non_countries)].copy()
df_cleaned = df_cleaned.sort_values(by="Country/Region").reset_index(drop=True)
non_states = [
    "Diamond Princess", "Grand Princess", "MS Zaandam",
    "Repatriated Travellers", "Unassigned",
    "French Guiana", "Greenland", "Hong Kong", "Macau",
    "Guadeloupe", "Martinique", "Faroe Islands", "Gibraltar",
    "Bermuda", "Isle of Man","Recovered"
]

# Drop rows where Province_State is in the non_states list
df_cleaned_2 = df_cleaned[~df_cleaned["Province/State"].isin(non_states)].copy()

df_cleaned_2.to_csv("data/time_series_covid19_confirmed_global.csv", index= False)
```

### 2. Data Formating

> Reformating the data columns and aggregate them in the same date

```python
df_melt = pd.melt(frame=df,
        id_vars=df.columns[:4],
        value_vars=df.columns[4:],
        var_name="Date",
        value_name="Confirmed"
)

df_melt.Date = pd.to_datetime(df_melt.Date)
```
